```python
# Convert all string values in the DataFrame to lowercase
df= df.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)

# Preview the result
print(df.head())
```

```python
#Standadization for Qualification

# Define normalization map 
qualification_map = {
    "phd": "phd",
    "doctorate": "phd",
    "master": "master's",
    "masters": "master's",
    "master's": "master's",
    "bachelor": "bachelor's",
    "bachelors": "bachelor's",
    "bachelor's": "bachelor's",
    "mba": "master's",
    "associate": "associate"
}

# Apply the Standadization
df["Qualification"] = df["Qualification"].apply(
    lambda x: qualification_map.get(x.strip().lower(), x.strip().lower())
)
```

```python
#Standaization for salary range 

# Function to extract and convert salary values
def parse_salary_range(s):
    try:
        parts = s.replace('$', '').replace(',', '').split('-')
        min_salary = int(parts[0].strip())
        max_salary = int(parts[1].strip())
        return pd.Series([min_salary, max_salary])
    except:
        return pd.Series([None, None])  # handle invalid formats

# Apply parsing
df[['Min Salary', 'Max Salary']] = df['Salary Range'].apply(parse_salary_range)
df['Avg Salary']=df[['Min Salary','Max Salary']].mean(axis=1)
# Optionally drop the original string column
# df_cleaned.drop(columns=['Salary Range'], inplace=True)
```
