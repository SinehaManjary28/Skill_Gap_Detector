```python
import pandas as pd
from ydata_profiling import ProfileReport

# Load the CSV
df = pd.read_csv("final_synthetic_skillsets.csv")

# Generate the profiling report
profile = ProfileReport(df, title="Skill_profile", explorative=True)

# Display report in a notebook (if using Jupyter or Colab)
profile.to_notebook_iframe()

# Or save it to an HTML file
profile.to_file("Skill_profile.html")
```

```python
# Convert all string values in the DataFrame to lowercase
df= df.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)

# Preview the result
df.head()
```

```python
import ast

# Step 1: Convert stringified lists to actual Python lists
df['Required Skills'] = df['Required Skills'].apply(ast.literal_eval)

# Step 2: Clean (lowercase and strip)
df['Required Skills'] = df['Required Skills'].apply(
    lambda skills: [skill.lower().strip() for skill in skills]
)

# Step 3: Skill mapping for standardization for skills col
skill_map = {
    'ms excel': 'excel',
    'microsoft excel': 'excel',
    'powerbi': 'power bi',
    'dashboarding': 'dashboards',
    'jupyter notebook': 'jupyter',
    'data viz': 'data visualization',
    'nlp': 'natural language processing',
    'ml': 'machine learning',
    # Add more mappings as needed
}

# Step 4: Standardize skills using the map
def standardize_skills(skill_list):
    return [skill_map.get(skill, skill) for skill in skill_list]

df['Required Skills'] = df['Required Skills'].apply(standardize_skills)
```

```python
# Convert list of skills to comma-separated string
df['Required Skills'] = df['Required Skills'].apply(lambda x: ', '.join(x) if isinstance(x,list)else x)
```

```python
# Generate the profiling report
profile = ProfileReport(df, title="Skill_profile", explorative=True)

# Display report in a notebook (if using Jupyter or Colab)
profile.to_notebook_iframe()

# Or save it to an HTML file
profile.to_file("Skill_profile.html")
```
